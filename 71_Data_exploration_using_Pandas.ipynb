```python
import pandas

df = pandas.read_csv("salaries_by_college_major.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
#Figure out if there are any missing or junk data in our dataframe
#Look for NaN (Not A Number) values in our dataframe
df.isna()
```

```python
#Check the last couple of rows in the dataframe:
df.tail()
```

```python
#Delete the Last Row
#The first way is to manually remove the row at index 50
#The second way is to simply use the .dropna() method from pandas
clean_df = df.dropna()
clean_df.tail()
```

```python
#Print the new dataframe:
clean_df.isna()
```

```python
#To access a particular column from a data frame we can use the square bracket notation, like so:
clean_df['Starting Median Salary']
```

```python
# To find the highest starting salary we can simply chain the .max() method
clean_df['Starting Median Salary'].max()
```

```python
# The highest starting salary is $74,300. But which college major earns this much on average? 
#For this, we need to know the row number or index so that we can look up the name of the major
#Lucky for us, the .idxmax() method will give us index for the row with the largest value
clean_df['Starting Median Salary'].idxmax()
```

```python
# To see the name of the major that corresponds to that particular row, we can use the .loc (location) property
clean_df['Undergraduate Major'].loc[43]
```

```python
#------------OR--------------------------
# Here we are selecting both a column ('Undergraduate Major') and a row at index 43, so we are retrieving the value of a particular cell. You might see people using the double square brackets notation to achieve exactly the same thing: 

clean_df['Undergraduate Major'][43]
```

```python
#If you don't specify a particular column you can use the .loc property to retrieve an entire row:
clean_df.loc[43]
```

```python
# What college major has the highest mid-career salary? How much do graduates with this major earn?
print(clean_df['Mid-Career Median Salary'].max())
print(clean_df['Mid-Career Median Salary'].idxmax())
print(clean_df['Mid-Career Median Salary'][8])
```

```python
clean_df['Mid-Career Median Salary'].max()
clean_df['Mid-Career Median Salary'].idxmax()
```

```python
# Which college major has the lowest starting salary and how much do graduates earn after university?
print(clean_df['Starting Median Salary'].min())
print(clean_df['Starting Median Salary'].idxmin())
clean_df['Undergraduate Major'].loc[clean_df['Starting Median Salary'].idxmin()]
```

```python
#How would we calculate the difference between the earnings of the 10th and 90th percentile? Well, Pandas allows us to do simple arithmetic with entire columns, 
#so all we need to do is take the difference between the two columns:
print(clean_df['Mid-Career 90th Percentile Salary'])
print(clean_df['Mid-Career 10th Percentile Salary'])

```

```python
clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
#-------or----------
#clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
```

```python
#The output of this computation will be another Pandas dataframe column. We can add this to our existing dataframe with the .insert() method:
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread', spread_col) #1 is the 1th index of column where we want this new column
clean_df.head()
```

```python
#To see which degrees have the smallest spread, we can use the .sort_values() method. 
#And since we are interested in only seeing the name of the degree and the major, we can pass a list of these two column names to look at the .head() of these two columns exclusively.
# .sort_values() sorts in ascending
low_risk = clean_df.sort_values('Spread')
print(low_risk)
print("-------------------------------------------")
print("-------------------------------------------")
print("-------------------------------------------")
print(low_risk[['Undergraduate Major', 'Spread']].head())
```

```python
# Find the top 5 degrees with the highest values in the 90th percentile.
print(clean_df["Mid-Career 90th Percentile Salary"])
print("--------------------------------------------------------------------------------")
print(clean_df["Mid-Career 90th Percentile Salary"].sort_values())
highest_potential = highest_potential = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False) #sets it to descending **
highest_potential[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()

```

```python
# Often times you will want to sum rows that belong to a particular category
# For example, which category of degrees has the highest average salary? Is it STEM, Business or HASS (Humanities, Arts, and Social Science)? 
# To answer this question we need to learn to use the .groupby() method. This allows us to manipulate data similar to a Microsoft Excel Pivot Table
# We have three categories in the 'Group' column: STEM, HASS and Business. Let's count how many majors we have in each category:
clean_df.groupby('Group').count()
```

```python
# Mini Challenge
# Now can you use the .mean() method to find the average salary by group?
clean_df.groupby('Group').mean()
```

```python
# Number formats in the Output
# The above is a little hard to read, isn't it? We can tell Pandas to print the numbers in our notebook to look like 1,012.45 with the following line:
pandas.options.display.float_format = '{:,.2f}'.format
clean_df.groupby('Group').mean()
```
